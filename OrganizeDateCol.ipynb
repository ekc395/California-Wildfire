```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load and view dataframe
fireData = pd.read_csv("cleanedCAFire.csv", index_col = "_id", low_memory = False)
fireData
```

```python
# See data types per col
print(fireData.dtypes)
```

```python
# Separate 'Incident Start Date' column into 'Day', 'Month', 'Year' and get rid of time
fireData['Incident Start Date'] = pd.to_datetime(fireData['Incident Start Date'])
fireData['Day'] = fireData['Incident Start Date'].dt.day
fireData['Month'] = fireData['Incident Start Date'].dt.month
fireData['Year'] = fireData['Incident Start Date'].dt.year
fireData
```

```python
# Check if added
print(fireData.columns)
```

```python
# Drop 'Incident Start Date'
fireData = fireData.drop(columns=['Incident Start Date'])
```
