```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/home/lxinzhui/py/Whelk-Repo/Project/cleanedCAFire.csv", index_col="_id", low_memory=False)
```

```python
df_grouped = df.groupby(["* City", "County"]).size().reset_index(name="Incident Count")

df_grouped = df_grouped.sort_values(by="Incident Count", ascending=False)
```

```python
df_filtered = df_grouped[df_grouped["Incident Count"] > 1000]

locations = df_filtered["* City"] + ", " + df_filtered["County"]
incident_counts = df_filtered["Incident Count"]

plt.figure(figsize=(12, 6))
plt.bar(locations, incident_counts)
plt.xticks(rotation=90, ha="right")
plt.xlabel("Location (City, County)")
plt.ylabel("Number of Incidents")
plt.title("Number of Fire Incidents by Location (More than 1000 Incidents)")
plt.show()
```
